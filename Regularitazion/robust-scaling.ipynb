```python
from sklearn.preprocessing import RobustScaler
import pandas as pd

# Supongamos que tienes un DataFrame de pandas llamado 'data' con las características a escalar
# Por ejemplo:
data = pd.DataFrame({'feature1': [10, 20, 30, 40, 50],
                     'feature2': [1, 2, 3, 100, 200]})

# Inicializar el escalador robusto
robust_scaler = RobustScaler()

# Escalar las características
scaled_data = robust_scaler.fit_transform(data)

# El resultado es un array numpy, así que puedes convertirlo de nuevo a DataFrame si lo necesitas
scaled_df = pd.DataFrame(scaled_data, columns=data.columns)

# Ahora 'scaled_df' contiene las características escaladas robustamente
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crear un histograma de las características antes del escalado
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.histplot(data, bins=10)
plt.title('Antes del escalado robusto')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

# Crear un histograma de las características después del escalado
plt.subplot(1, 2, 2)
sns.histplot(scaled_df, bins=10)
plt.title('Después del escalado robusto')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
